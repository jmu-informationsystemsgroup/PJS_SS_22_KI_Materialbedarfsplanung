```python
import numpy as np
import cv2
import os
from pathlib import Path

# Google Drive in Colab einbinden
#from google.colab import drive
#drive.mount('/content/drive')
```

```python
# Quell- und Zielpfade angeben
srcpath = "C:/Users/DanielWeller/Desktop/Testdaten/"
destpath = "C:/Users/DanielWeller/Desktop/bearbeitet/"


# An welcher Stelle des Dateinamens soll der Inhalt Ã¼berschrieben werden.
# Achtung ZÃ¤hler beginnt bei 0!
labelstelle = 4

# Mit welchem Labelflag soll der neue Dateinamen erzeugt werden
labelinhalt = 1000


# Iteration Ã¼ber alle Dateien des Ordners
for filename in os.listdir(srcpath):

    # Erzeugt Filename ohne Endung. Aus Testname.jpg wird Testname
    filenameshort = Path(filename).stem
    print("Alter Filename          : "+str(filenameshort))

    # Splitted den Dateinamen mit dem Trennoperator "_" und speichert den Inhalt in eine Liste
    splittedfilename = filenameshort.split("_")
    #print("Splitted Filename: "+str(splittedfilename))
    
    # Ãberschreibt den alten Labelinhalt mit dem neuen Labelinhalt der Variable "labelinhalt" an der Stelle "labelstelle"
    splittedfilename[labelstelle] = str(labelinhalt)
    #print("Splitted Filename mit neuem Labelinhalt: "+str(splittedfilename))

    # Joined den Inhalt der Liste "splittedfilename" zum neuen Filename
    newfilename = "_".join(splittedfilename)
    print("Neuer Filename nach join: "+str(newfilename))

    # LÃ¤dt das img aus Pfad des Filenames
    img = cv2.imread(os.path.join(srcpath,filename))






    ###########################################################
    # Bildbearbeitung an dieser Stelle
    ###########################################################








    # Speichern des neuen Bildes im Zielpfad
    cv2.imwrite(os.path.join(destpath, newfilename+'.jpg'),img)
```
