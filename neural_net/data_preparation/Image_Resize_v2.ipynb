```python
import numpy as np
import cv2
import os
from pathlib import Path
```

```python
# Kleines Hilfsskript um die Bilddaten welche sich in einem Ordner befinden im Anschluss 
# auf eine Standarddisierte Auflsung zu resizen

# Quell- und Zielpfade angeben
srcpath = "C:/Users/DanielWeller/Desktop/renamed/"
destpath = "C:/Users/DanielWeller/Desktop/resized/"

# Zuknftige Mae fr Querformat festlegen 
widthquer = 4000
heightquer = 3000

# Zuknftige Mae fr Hochformat festlegen 
widthhoch = 3000
heighthoch = 4000

# An welcher Stelle des Dateinamens soll der Inhalt berschrieben werden.
# Achtung Zhler beginnt bei 0!
labelstelle = 12
# Welche Ausprgung soll an der vorher gewhlten Labelstelle eingefgt werden.
labelinhalt = 1

# Fr das Resizen ist das Aufnahmeformat Hoch-/Querformat wichtig. An welcher Stelle des originalen Dateinamens befindet sich der Labelflag fr Hoch/Querformat.
# Achtung Zhler beginnt bei 0!
labelstelleAufnahmeformat = 14
# Welche Ausprgung entspricht Querformat im originalen Dateinamen:
quer = 1
# Welche Ausprgung entspricht Hochformat im originalen Dateinamen:
hoch = 2

# Iteration ber alle Dateien des Ordners
for filename in os.listdir(srcpath):

    # Erzeugt Filename ohne Endung. Aus Testname.jpg wird Testname
    filenameshort = Path(filename).stem
    print("Alter Filename          : "+str(filenameshort))

    # Splitted den Dateinamen mit dem Trennoperator "_" und speichert den Inhalt in eine Liste
    splittedfilename = filenameshort.split("_")
    #print("Splitted Filename: "+str(splittedfilename))
    
    # berschreibt den alten Labelinhalt mit dem neuen Labelinhalt der Variable "labelinhalt" an der Stelle "labelstelle"
    splittedfilename[labelstelle] = str(labelinhalt)
    #print("Splitted Filename mit neuem Labelinhalt: "+str(splittedfilename))

    # Joined den Inhalt der Liste "splittedfilename" zum neuen Filename
    newfilename = "_".join(splittedfilename)
    print("Neuer Filename nach join: "+str(newfilename))

    # Ldt das img aus Pfad des Filenames
    img = cv2.imread(os.path.join(srcpath,filename))

    # Prfen ber "LabelstelleAufnahmeformat" ob original Img im Hoch- oder Querformat aufgenommen wurde.
    if splittedfilename[labelstelleAufnahmeformat] == str(quer):
        dim = (widthquer, heightquer)
    elif splittedfilename[labelstelleAufnahmeformat] == str(hoch):
        dim = (widthhoch, heighthoch)

    # Bild wird mit neuer Hhe und Weite versehen
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

    # Speichern des neuen Bildes im Zielpfad
    cv2.imwrite(os.path.join(destpath , newfilename+'.jpg'),resized)
```
