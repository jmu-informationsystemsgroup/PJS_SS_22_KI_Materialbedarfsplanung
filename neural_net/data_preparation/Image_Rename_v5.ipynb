```python
import cv2
import os
from pathlib import Path
import shutil
import numpy as np
```

```python
# Kleines Hilfsskript um die Bilddaten (roh) welche in Unterordner einsortiert sind, in 
# einen Dateiordner zu kopieren mit standardisierter Dateinamensstruktur. Dieser Prozess ist nach jedem Baustellen-
# besuch erforderlich.

# Quell- und Zielpfade:
srcpath = "C:/Users/DanielWeller/Desktop/Bilddaten roh/"
destpath = "C:/Users/DanielWeller/Desktop/renamed/"

# bersicht ber die Dateinamenstruktur:
# Trennoperator ist "_"
#
# Stelle:         Bedeutung:
# 0               UserID_
# 1               BaustellenID
# 2               WandID
# 3               BildID
# 4               Synthetisch - Spiegeln
# 5               Synthetisch - Rotation
# 6               Synthetisch - Noise
# 7               Platzhalter4
# 8               Synthetisch - Helligkeit
# 9               Platzhalter6
# 10              Synthetisch - Schrfe
# 11              Platzhalter8
# 12              Platzhalter9        
# 13              Aufnahmegert(DSLR/Smartphone)
# 14              Aufnahmeformat(Quer/Hoch)
# 15              Stickerart(rot, Aruco etc.)



# Variablen fr die Bennenung (1 steht fr Originalbild)
userid = 1
platzhalter1 = 1
platzhalter2 = 1
platzhalter3 = 1
platzhalter4 = 1
platzhalter5 = 1
platzhalter6 = 1
platzhalter7 = 1
platzhalter8 = 1
platzhalter9 = 1

# Erzeugt zuerst ein leeres Array. Anpassung der Array Lnge kann erforderlich werden. Festlegen welche Ausprgung welche Aufnahmeart bekommt
aufnahmegeraet = [None] * 10
aufnahmegeraet[1] = 'Smartphone'
aufnahmegeraet[2] = 'DSLR'

# Erzeugt zuerst ein leeres Array. Anpassung der Array Lnge kann erforderlich werden. Festlegen welche Ausprgung welches Aufnahmeformat bekommt
aufnahmeformat = [None] * 10
aufnahmeformat[1] = 'Quer'
aufnahmeformat[2] = 'Hoch'

# Erzeugt zuerst ein leeres Array. Anpassung der Array Lnge kann erforderlich werden. Festlegen welche Ausprgung welche Stickerart bekommt
stickerart = [None] * 10
stickerart[1] = 'ohne'
stickerart[2] = 'mitAufkleberRot'
stickerart[3] = '3Aruco'
stickerart[4] = '1Aruco'




# Legt die Rahmenparameter fr die Iterationen fest. Es gilt unntige Iterationen zu vermeiden.
# 
# 
hoechstanzahlBaustellen = 9
hoechstanzahlWaendeProBaustelle = 70


# For-Schleife iteriert ber die Anzahl der Baustellen.
for counterBaustelle in range(1, hoechstanzahlBaustellen+1):

  # For-Schleife iteriert ber die Anzahl der Wnde/Ordner.
  for counterWand in range(1, hoechstanzahlWaendeProBaustelle+1):
      
      # Counter fr die Anzahl der Fotos im Dateinamen
      counterBild=1

      # Aktueller Pfad
      actpath = str(srcpath)+"Baustelle"+str(counterBaustelle)+"/Wand"+str(counterWand)

      # Prfen ob der Pfad tatschlich existiert, da ber Baustellen und Wnde einfach iteriert wird und es somit sein kann das der Pfad gar nicht existiert.
      if os.path.exists(actpath):
        actpathSavedWand = actpath
        # print("Aktueller Pfad der gespeichert wurde: "+str(actpathSavedWand))

        # For-Schleife iteriert ber die unterschiedlichen Aufnahmegerte
        for filename in os.listdir(actpath):
          aktuellesAufnahmegeraet = filename
          actpath = str(actpath)+"/"+str(aktuellesAufnahmegeraet)
          # print("Aktueller Pfad Aufnahmegerte: "+str(actpath))
          # print(str(os.path.isfile))
          if os.path.isfile(actpath):
            actpath = actpathSavedWand
            # print("Aktueller Pfad der von if Bedingung zurckgeschrieben wurde: "+str(actpath))
            continue
          # print("Aktueller Pfad vor dem Speichern in die Variable actpathSavedAufnahmegeraet: "+str(actpath))
          actpathSavedAufnahmegeraet = actpath
          # print("Aktueller Pfad der in Variable actpathSavedAufnahmegeraet gespeichert wurde: "+str(actpathSavedAufnahmegeraet))

          # For-Schleife iteriert ber die unterschiedlichen Aufnahmeformate
          for filename in os.listdir(actpath):
            # print("Aktueller Filename: "+str(filename))
            aktuellesAufnahmeformat = filename
            actpath = str(actpath)+"/"+str(aktuellesAufnahmeformat)
            # print("Aktueller Pfad Aufnahmeformat: "+str(actpath))
            if os.path.isfile(actpath):
              actpath = actpathSavedAufnahmegeraet
              # print("Aktueller Pfad der von if Bedingung zurckgeschrieben wurde: "+str(actpath))
              continue
            actpathSavedAufnahmeformat = actpath
            # print("Aktueller Pfad der in Variable actpathSavedAufnahmeformat gespeichert wurde: "+str(actpathSavedAufnahmeformat))

            # For-Schleife iteriert ber die unterschiedlichen Stickerarten
            for filename in os.listdir(actpath):
              aktuelleStickerart = filename
              actpath = str(actpath)+"/"+str(aktuelleStickerart)
              # print("Aktueller Pfad Stickerart: "+str(actpath))
              if os.path.isfile(actpath):
                actpath = actpathSavedAufnahmeformat
                # print("Aktueller Pfad der von if Bedingung zurckgeschrieben wurde: "+str(actpath))
                continue
              actpathSavedStickerart = actpath
              # print("Aktueller Pfad der in Variable actpathSavedStickerart gespeichert wurde: "+str(actpathSavedStickerart))
              
              # For-Schleife iteriert durch die Dateien im jeweiligen Ordner
              for filename in os.listdir(actpath):
                # Legt den Quelldateipfad fest
                src_file = str(actpath)+"/"+str(filename)
                print("Quelldatei: "+str(src_file))
                # Legt den Zieldateipfad fest
                dest_file = (
                str(destpath)
                +str(userid)+"_"
                +str(counterBaustelle)+"_"
                +str(counterWand)+"_"
                +str(counterBild)+"_"
                +str(platzhalter1)+"_"
                +str(platzhalter2)+"_"
                +str(platzhalter3)+"_"
                +str(platzhalter4)+"_"
                +str(platzhalter5)+"_"
                +str(platzhalter6)+"_"
                +str(platzhalter7)+"_"
                +str(platzhalter8)+"_"
                +str(platzhalter9)+"_"
                +str(aufnahmegeraet.index(aktuellesAufnahmegeraet))+"_"
                +str(aufnahmeformat.index(aktuellesAufnahmeformat))+"_"
                +str(stickerart.index(aktuelleStickerart))
                +".jpg" )
                print("Zieldatei: "+str(dest_file))
                # Kopiert von Quell- zu Zieldatei
                shutil.copy2(src_file, dest_file)
                # Erhht den Dateinamen Counter um 1
                counterBild += 1
                # actpath = actpathSavedStickerart
                # # print("Aktueller Pfad der zurckgeschrieben wurde: "+str(actpath))
              actpath = actpathSavedAufnahmeformat
              # print("Aktueller Pfad der zurckgeschrieben wurde: "+str(actpath))
            actpath = actpathSavedAufnahmegeraet
            # print("Aktueller Pfad der zurckgeschrieben wurde: "+str(actpath))
          actpath = actpathSavedWand
          # print("Aktueller Pfad der zurckgeschrieben wurde: "+str(actpath))





```
